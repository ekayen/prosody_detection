Load the stuff:

```python
import pandas as pd
import numpy as np
```

```python
datafile = ('data/utterances.txt')
df = pd.read_csv(datafile,sep='\t',header=None)
```

```python
labels = df[3].tolist()
labels
keys = df[1].tolist()
keys = np.array(keys)
keys
```

Get a list of the labels

```python
last = [int(line[-1]) for line in labels]
last = np.array(last)
last
```

Count the zeros and the ones

```python
total = last.shape[0]
total_one = np.sum(last)
total_zero = total - total_one
```

```python
percentage0 = total_zero/total
percentage0
```

```python
percentage1 = total_one/total
percentage1
total_one
```

```python
total
```

Check the lengths

```python
lens = [len(line.split()) for line in labels]
lens = np.array(lens)
lens
```

```python
lens0 = lens[last==0]
lens0.shape
np.sum(lens0==1)/lens0.shape[0]
```

```python
lens1 = lens[last==1]
lens1.shape
np.sum(lens1==1)/lens1.shape[0]
```

Make a mask to filter out utterances of just one token

```python
unshorts = last[lens>1]
one_unshorts = np.sum(unshorts)
total_unshorts = unshorts.shape[0]
```

```python
one_unshorts/total_unshorts
```

Make a mask is `True` for only half of zeros

```python
rand_mask = np.random.rand(lens.shape[0]) > 0.5
zero_mask = last==0
rand_zero_mask = np.logical_and(rand_mask,zero_mask)
np.sum(zero_mask)
np.sum(rand_mask)
np.sum(rand_zero_mask)
```

`or` this with a mask that is `True` for all ones:

```python
one_mask = last==1
np.sum(one_mask)+np.sum(zero_mask)
keep = np.logical_or(rand_zero_mask,one_mask)
np.sum(keep)
```

```python
skewed = keys[keep]
skewed

len(skewed)
```

```python
set(skewed.tolist())
```

```python
import pickle
with open('skewed_label_list.pkl','wb') as f:
    pickle.dump(skewed,f)
```

Make a dictionary with the skewed dataset

```python
with open('data/cmvn_tensors.pkl','rb') as f:
    cmvn_dict = pickle.load(f)
```

```python
skewed_dict = {}
for tens in cmvn_dict:
    if tens in skewed:
        skewed_dict[tens] = cmvn_dict[tens]
```

```python
len(skewed_dict)
```

```python
with open('data/skewed_cmvn.pkl','wb') as f:
    pickle.dump(skewed_dict,f)
```

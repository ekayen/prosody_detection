Double check the Kaldi output that we've pickled to make sure that the time dimension makes sense given the number of seconds in the audio

```python
import pickle
import numpy as np
import torch
```

Load all the utterance feats that we've pickled as numpy arrays stored in a dictionary

```python
dict_file = 'data/utterances_feats.pkl'

with open(dict_file,'rb') as f:
     kaldi_dict = pickle.load(f)
    
```

Pull out the keys

```python
keys = sorted(list(kaldi_dict.keys()))
```

Use the timestamps which are baked into the keys to figure out how much time is spanned by the utterance. Save as another dictionary, where keys are the same keys, and values are the number of seconds spanned.

```python
time_dict = {}
for key in keys:
    times = [float(t) for t in key.split('_')[-1].split('-')]
    span = (times[1] - times[0])/100
    time_dict[key] = span
    
keys[98],time_dict[keys[98]]
```

Now go through all the keys and compare the number of seconds and the length of the features

```python
for key in keys:
    seconds = time_dict[key]
    frames = kaldi_dict[key].shape[0]
    print(seconds,frames)
```
